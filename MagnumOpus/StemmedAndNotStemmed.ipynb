Take one (big) document
1. Stem, Vectorise, Compute similarity
2. Don't Stem, Vectorise, Compute similarity

### Document


```python
abstract = "D:\\projects\\nlp-hands-on\\narutoWiki.txt"
internet1 = "D:\\projects\\nlp-hands-on\\internet.txt"
internet2 = "D:\\projects\\nlp-hands-on\\internet2.txt"

file1 = open(internet1, "r", encoding="utf8")
text1 = file1.read()

file2 = open(internet2, "r", encoding="utf8")
text2 = file2.read()
```

#### Tokenizing

```python
from nltk.tokenize import word_tokenize
tokens1 = word_tokenize(text1)
tokens2 = word_tokenize(text2)
```

```python
from nltk.stem import LancasterStemmer
lancaster = LancasterStemmer()
stemmed1 = [lancaster.stem(token) for token in tokens1]
stemmed2 = [lancaster.stem(token) for token in tokens2]
not_stemmed1 = tokens1
not_stemmed2 = tokens2
```

```python
# Vectorizing
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(binary=True)
corpus = ["One of the most basic ways we can numerically represent words"]
vectorizer.fit(corpus)
vector_stemmed1 = vectorizer.transform(stemmed1).toarray()
vector_stemmed2 = vectorizer.transform(stemmed2).toarray()
vector_stemmed1
```

```python
# Similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector_stemmed1,vector_stemmed2)
similarity
```
